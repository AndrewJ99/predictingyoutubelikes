### __Str lengths (`title_len` | `channelTitle_len` | `description_len`)__

```python
def get_str_length(description):
    return len(description) if description else 0
```

```python
# getting the string length of each row in training set
title_len = OUR_MAIN_df['title'].apply(get_str_length)
channelTitle_len = OUR_MAIN_df['channelTitle'].apply(get_str_length)
description_len = OUR_MAIN_df['description'].apply(get_str_length)
```

```python
# adding new columns to training set
OUR_MAIN_df['title_len'] = title_len
OUR_MAIN_df['channelTitle_len'] = channelTitle_len
OUR_MAIN_df['description_len'] = description_len
```

### __# of tags (`tags_count`)__

```python
# checks if tags are empty
def tagsEmpty(tags_list):
    return len(tags_list) == 1 and '[None]' in tags_list[0]

# count the number of tags
def tagsCount(tags):
    tags_list = tags.split('|')
    return 0 if tagsEmpty(tags_list) else len(tags_list)
```

```python
# gets the number of tags
tags_count = OUR_MAIN_df['tags'].apply(tagsCount)
```

```python
# adds new column to training & testing set
OUR_MAIN_df['tags_count'] = tags_count
```

# __Random Forest Regression__

```python
def scoring_RFR(X_train, X_test, y_train, y_test, dataset_name):
    arr = [10, 50, 100]
    RsquareArr = []
    MAEarr = []
    for x in arr:
        RFreg = RandomForestRegressor(n_estimators = x, random_state = 0)
        RFreg.fit(X_train, y_train)
        y_predict = RFreg.predict(X_test)
        
        r_square = r2_score(y_test, y_predict)
        MAE = mean_absolute_error(y_test, y_predict)
        
        RsquareArr.append(r_square)
        MAEarr.append(MAE)

    print(RsquareArr)
    print(MAEarr)

    print('\n')
    
    print(f'{dataset_name} Dataset - R_Squared w/ n_estimators 10, 50, 100')
    print('----------------------------------------------------')
    for i in range(3):
        print(f'R_Squared = {RsquareArr[i]:.3f}','%')
        
    print(f'{dataset_name} Dataset - MAE w/ n_estimators 10, 50, 100')
    print('----------------------------------------------------')
    for i in range(3):
        print(f'MAE = {MAEarr[i]:.3f}')
```

# __LinearSVR__

```python
def act_n_pred2(y_act, y_pred, n=10):
    print(                 f'{r2_score(y_act, y_pred).round(3):>10} : R^2 (The best possible score is 1.0, lower values are worse.)')
    print(      f'{mean_absolute_error(y_act, y_pred).round(3):>10} : Mean Absolute Error')
```

### Main

```python
df = OUR_MAIN_df.copy()

# training dataset
train_X = df.loc[:, df.columns != 'likes']
train_y = pd.DataFrame(df['likes'])
```

```python
#Features Scaled
X = preprocessing.scale(train_X)
X_df = pd.DataFrame(X)

#Splitting
X_train, X_test, y_train, y_test = train_test_split(X_df, train_y, test_size=0.3, random_state=3)
```

```python
lsvr = LinearSVR()

lsvr.fit(X_train, y_train)
Y_preds = lsvr.predict(X_test)
```

```python
print(Y_preds[:10])
print(y_test[:10])
```

```python
print(f'Test R^2     : {r2_score(y_test, Y_preds):.3f}')
print(f'Test R^2     : {lsvr.score(X_test, y_test):.3f}')
print(f'Training R^2 : {lsvr.score(X_train, y_train):.3f}')
```

```python
grid = GridSearchCV(LinearSVR(),param_grid = {'C': [1.0, 0.1, 0.01, 10.0]}, scoring="r2", cv=5)
grid.fit(X_train, y_train)

print('Best Parameters : ',grid.best_params_)
print('Best Score      : ',grid.best_score_)
print('Test R^2        : %.3f'%r2_score(y_test, grid.best_estimator_.predict(X_test)))
print('Test R^2        : %.3f'%grid.best_estimator_.score(X_test, y_test))
print('Training R^2    : %.3f'%grid.best_estimator_.score(X_train, y_train))

Y_preds = grid.best_estimator_.predict(X_test)
```

```python
act_n_pred2(y_test, Y_preds)
```
